```python
#pivot data
import pandas as pd
```

```python

data={'date':['5/01/2025','5/02/2025','5/03/2025','5/01/2025','5/02/2025','5/03/2025','5/01/2025','5/02/2025','5/03/2025'],
      'city':['newyork','newyork','newyork','mumbai','mumbai','mumbai','beijng','beijng','beijng'],
      'temperature':[65,66,68,75,78,82,80,77,79],
      'humidity':[56,58,60,80,83,85,26,30,35]}
df=pd.DataFrame(data)
df
```

```python
df.pivot(index="date",columns="city")
```

```python
df.pivot(index="temperature", columns="date",values="humidity")
```

```python
#pivot tables it is used to summarize/aggregate data
data={'date':['5/01/2025','5/01/2025','5/02/2025','5/02/2025','5/01/2025','5/01/2025','5/02/2025','5/02/2025'],
      'city':['newyork','newyork','newyork','newyork','mumbai','mumbai','mumbai','mumbai'],
      'temperature':[65,61,70,72,75,78,82,80],
      'humidity':[56,54,60,62,80,83,85,26]}
df=pd.DataFrame(data)
df
```

```python
df.pivot_table(index="city",columns="date",aggfunc="sum")

#by using aggfunc we can sum,count,mean diff values in a common
```

```python
df.pivot_table(index="city",columns="date",margins="True")
#by this we can know the all rows and columns which gives average of the details
```

```python

data={'date':['5/01/2025','5/02/2025','5/03/2025','13/01/2025','13/02/2025','13/03/2025'],
      'city':['newyork','newyork','newyork','newyork','newyork','newyork'],
      'temperature':[65,66,68,24,23,20],
      'humidity':[56,58,60,80,83,85]}
df=pd.DataFrame(data)
df
```

```python

df2=df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
df2
#in this we can get the average of 2 days in a month
```

```python
#melt
data={'date':['5/01/2025','5/02/2025','5/03/2025','13/01/2025','13/02/2025','13/03/2025'],
      'city':['newyork','newyork','newyork','newyork','newyork','newyork'],
      'temperature':[65,66,68,24,23,20],
      'humidity':[56,58,60,80,83,85]}
df=pd.DataFrame(data)
df
```

```python
df.melt(id_vars='temperature')
#by using melt we will be getting it in detail
```

```python
#data merging & joining
import pandas as pd
```

```python
df1=pd.read_csv('/content/sample_data/LOTR.csv')
df1
```

```python
df2=pd.read_csv('/content/sample_data/LOTR 2.csv')
df2
```

```python
  df1.merge(df2)
  #by default it gives common values
```

```python
df1.merge(df2,how='inner', on=['FellowshipID','FirstName'])
```

```python
df1.merge(df2,how='outer')
#it gives all the values in both the data sets taking common values in them
```

```python
df1.merge(df2,how='left')
#it just gives left side data given and excludes right
```

```python
df1.merge(df2,how='right')
#it just gives right side data given and excludes left
```

```python
df1.merge(df2,how='cross')
# it takes each value from left dataframe and compares to right dataframe
```

```python
#join
df1.join(df2,on='FellowshipID',how='outer',lsuffix='_left',rsuffix='_right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_left',rsuffix='_right')
df4
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_left',rsuffix='_right',how='outer')
df4
```

```python
#concatinate
pd.concat([df1,df2],join='outer',axis=1)
```

```python
#append -this method was deleted
df5=df1.append(df2)
df5
```
