```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Bengaluru_House_Data.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull()
```

```python
df2=df.fillna(value=0) # the value given in value will get replaced with null values
df2
```

```python
df2.isnull().sum().sum()
```

```python
df.isnull().sum().sum()
```

```python
# finning null values with previous value
df3=df.fillna(method='pad') #using pad method can fill the previous val
df3
```

```python
df3.isnull().sum().sum()
# it is a concise way to count all the NaN values present in the DataFrame df3
```

```python
d4=df.fillna(method='bfill') #using bfill method can fill the next value
d4
```

```python
d5=df.fillna(method='pad',axis=1) #using pad method can fill the previous val
d5
```

```python
d6=df.fillna(method='bfill',axis=1) #using bfill method can fill the next value
d6
```

```python
#filling diff values in  null with diff columns in this we've filled abc in society and balcony columns null values
df7 = df.fillna({'society':'abc','balcony':'xyz'})
df7
```

```python
#to fill null values with mean of the column
df8=df.fillna(value=df['balcony'].mean())
df8
```

```python
#to fill null values with max of the column
df9=df.fillna(value=df['balcony'].max())
df9
```

```python
df10=df.dropna()
df10
```

```python
df11=df.dropna(how='any')
df11
```

```python
#to replace the null values
import numpy as np
df12=df.replace(to_replace=np.nan, value=123)
df12
```

```python
df13=df.replace(to_replace=3.0, value=456)
df13
#we can easily replace the numerical values too
```

```python
#interpolate it checks with the data set and interpolates it to fill the null values
df['balcony']=df['balcony'].interpolate(method='linear')
df
```

```python
#interpolate it checks with the data set and interpolates it to fill the null values
df['balcony']=df['balcony'].interpolate(method='linear',limit_direction='forward')
df
```
